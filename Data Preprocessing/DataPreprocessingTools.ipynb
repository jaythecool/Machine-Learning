Importing the libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing the dataset

```python
dataset = pd.read_csv("Data.csv")
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

Taking care of missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
imputer.fit(X[:,1:3])
X[:,1:3] = imputer.transform(X[:,1:3])
```

```python
print(X)
```

Encoding categorical columns

Encoding independent variables

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# the class ColumnTransformer takes two parameters transformers and remainder
# transformers parameter expects a list of tuples where the first element is the type of transformer,
# second element is the class with which we want to encode the columns and the last one is the indices of
# the columns we want to encode, and in the remiander parameter we pass the value passthrough which indecates that
# we want to keep the original columns

ct = ColumnTransformer(transformers=[("encoder", OneHotEncoder(), [0])], remainder = "passthrough")

X = np.array(ct.fit_transform(X)) #needs to be converted into np array for further model training
```

```python
print(X)
```

Encoding dependent variables

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

y = le.fit_transform(y)
```

```python
print(y)
```

Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

Feature Scaling

```python
# we always apply feature scaling after splitting our dataset into trainng and test sets to avoid
# information leaking which could happen if we apply feature scaling before spliting the dataset as when we
# apply feature scaling before spliting the dataset it will capture the mean and standard daviation of all the data
# and thus our machine learning model will start neglecting values if they are not in one scale which will make our machine learning model less effecient

```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,3:] = sc.fit_transform(X_train[:,3:])
X_test[:,3:] = sc.transform(X_test[:,3:])
```
